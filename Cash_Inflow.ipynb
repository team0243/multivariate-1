```python
import matplotlib.pyplot as plt
```

```python
y = [9400,6890,4000,2600]
x = ['30/12/21', '4/01/22', '10/01/22', '14/01/22']
plt.bar(x,y,color='#dddddd' , alpha=0.8, hatch='\\\\')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.style.use('seaborn')

y = (9400,6890,4000,2600)
x = ['30/12/21', '4/01/22', '10/01/22', '14/01/22']

plt.title('Cash Inflow', fontname='tahoma',color='blue')

plt.ylabel('earnings(Bath)',color='m',fontname='tahoma',fontsize=14)
plt.xlabel('date',color='g',fontname='tahoma',fontsize=14)

plt.yticks(color='r')
plt.xticks(rotation='vertical',color='r',fontname='tahoma',fontsize=13)


plt.plot(x,y)
plt.show()
```
